# Statistical plotting with Seaborn (regression plots, distribution plots)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Loading Datasets

```python
df = sns.load_dataset("penguins")
df.head()
```

# 1. Linear Regression Plot (sns.regplot)

In Seaborn (sns), `regplot` is a function that creates a regression plot showing the relationship between two variables.

### Here's a brief overview of what `regplot` does:

- It takes two variables, x and y, as input.
- It calculates the regression line between the two variables using a linear model.
- It plots the data points for the two variables on a scatterplot.
- It adds the regression line to the plot, showing the predicted values based on the linear model.
- It also displays the confidence interval for the regression line (optional).

By default, `regplot` uses a simple linear regression model, but you can also specify other models, such as a logistic regression or a robust regression, using the `model` parameter.


```python
sns.regplot(data = df, x = "flipper_length_mm", y = "bill_depth_mm")
plt.show()
```

# 2. Multiple Linear Regression Plot (sns.lmplot)

In Seaborn (sns), `lmplot` is a high-level function for drawing a linear regression model across facets of a dataset. It combines `FacetGrid` with `regplot`, providing a more powerful interface for plotting linear models.

### Here's a brief overview of what `lmplot` does:

- It takes two variables, x and y, as input.
- It calculates the regression line between the two variables using a linear model.
- It plots the data points for the two variables on a scatterplot.
- It adds the regression line to the plot, showing the predicted values based on the linear model.
- It also displays the confidence interval for the regression line (optional).
- It allows for the creation of multiple subplots based on the values of other variables (faceting).

### Key Features:

- **Faceting:** You can create multiple plots based on the values of another variable using the `col`, `row`, and `hue` parameters.
- **DataFrame Input:** `lmplot` is designed to work well with DataFrames, allowing you to pass in a DataFrame directly and specify the variables by column names.
- **Customization:** You can customize the appearance of the plot using various parameters, such as `markers`, `paletttal_bill', y='tip', data=df)


```python
sns.lmplot(data = df, x = "flipper_length_mm", y = "bill_depth_mm", hue = "sex")
plt.show()
```

# 3. Residual Plot (sns.residplot


In Seaborn (sns), `residplot` is a function that creates a residual plot, which is used to check the goodness of fit for a regression model by displaying the residuals between observed and predicted values.

### Here's a brief overview of what `residplot` does:

- It takes two variables, x and y, as input.
- It fits a regression model and calculates the residuals (differences between observed and predicted values).
- It plots the residuals on the y-axis and the independent variable on the x-axis.

### Key Features:

- **Residuals:** Residuals are the differences between the observed values and the values predicted by the regression model. They indicate how well the model fits the data.
- **Visualization:** A residual plot helps visualize whether the residuals are randomly distributed or if there is a pattern, which can indicate issues with the model fit.
- **Linear and Non-Linear Models:** While primarily used for linear regression models, `residplot` can also be used to check the fit of non-tip', data=df)
plt.show()


```python
sns.residplot(data = df, x = "flipper_length_mm", y = "bill_depth_mm")
plt.show()
```

# 4. Histogram (sns.histplot)

- A histogram is a classic visualization tool that represents the distribution of one or more variables by counting the number of observations that fall within discrete bins.
- 
This function can normalize the statistic computed within each bin to estimate frequency, density or probability mass, and it can add a smooth curve obtained using a kernel density estimate, similar t`o kdeplot(`).

```python
sns.histplot(data = df, x = "flipper_length_mm")
plt.show()
```

# 5. Kernel Density Estimate (KDE) Plot (sns.kdeplot)
A kernel density estimate (KDE) plot is a method for visualizing the distribution of observations in a dataset, analogous to a histogram. KDE represents the data using a continuous probability density curve in one or more dimensions.

The approach is explained further in the user guide.

Relative to a histogram, KDE can produce a plot that is less cluttered and more interpretable, especially when drawing multiple distributions. But it has the potential to introduce distortions if the underlying distribution is bounded or not smooth. Like a histogram, the quality of the representation also depends on the selection of good smoothing parameters.

```python
sns.kdeplot(data = df, x = "flipper_length_mm", y = "bill_depth_mm", hue = "sex")
plt.show()
```

# 6. Distribution Plot (sns.distplot)

In Seaborn (sns), `distplot` is a function that combines a histogram and a kernel density estimate (KDE) to visualize the distribution of a dataset.

### Here's a brief overview of what `distplot` does:

- It takes a single variable as input and plots its distribution.
- It combines a histogram and a kernel density estimate to visualize the distribution of the data.
- It can optionally fit a parametric distribution to the data and plot that as well.

### Key Features:

- **Histogram:** A histogram is a bar plot where each bar represents the frequency or count of data points in a bin.
- **Kernel Density Estimate (KDE):** KDE is a smoothed version of the histogram that represents the data using a continuous probability density curve.
- **Parametric Distribution:** You can fit and plot a parametric distribution to the data using the `fit` parameter.

```python
sns.distplot(x = df["flipper_length_mm"], hist = True, kde = True, rug = True)
plt.show()
```

# 7. Box Plot (sns.boxplot)
- A box plot (or box-and-whisker plot) shows the distribution of quantitative data in a way that facilitates comparisons between variables or across levels of a categorical variable. The box shows the quartiles of the dataset while the whiskers extend to show the rest of the distribution, except for points that are determined to be outliers using a method that is a function of the inter-quartile range.

```python
sns.boxplot(data = df.head(100), x = "flipper_length_mm", hue = "sex")
plt.show()
```

# 8. Violin Plot (sns.violinplot)
- A violin plot plays a similar role as a box-and-whisker plot. It shows the distribution of data points after grouping by one (or more) variables. Unlike a box plot, each violin is drawn using a kernel density estimate of the underlying distribution.

```python
df = sns.load_dataset("titanic")
#sns.violinplot(x=df["age"])
sns.violinplot(data = df.head(100), x = "age", hue = "sex")
plt.show()
```

# 9. Pair Plot (sns.pairplot)
- Plot pairwise relationships in a dataset.
-

The diagonal plots are treated differently: a univariate distribution plot is drawn to show the marginal distribution of the data in each column.

It is also possible to show a subset of variables or plot different variables on the rows and columns.

This is a high-level interfac`e for Pa`irGrid that is intended to make it easy to draw a few common styles. You shoul`d use Pa`irGrid directly if you need more flexibility.

```python
#sns.violinplot(x=df["age"])
sns.pairplot(data = df.head(100), hue = "sex")
plt.show()
```

# 10. Heatmap (sns.heatmap)
Plot rectangular data as a color-encoded matrix.

This is an Axes-level function and will draw the heatmap into the currently-active Axes if none is provided to the ax argument. Part of this Axes space will be taken and used to plot a colormap, unless cbar is False or a separate Axes is provided t`o cbar_a`
`Mostly used in Confusion Matrix`x.

```python
data = np.random.rand(4, 4)
columns = list('ABCD')
index = list('abcd')
df = pd.DataFrame(data, columns=columns, index=index)

sns.heatmap(data = df, annot = True)
plt.show()
```

# 11. Joint Plot (sns.jointplot)
This function provides a convenient interface to the JointGrid class, with several canned plot kinds. This is intended to be a fairly lightweight wrapper; if you need more flexibility, you should use JointGrid directly.

```python
df = sns.load_dataset("titanic")
plt.figure(figsize = (15,8))
sns.jointplot(data = df.head(100), x = "age", y = "pclass", hue = "sex")
plt.show()
```

# 12. Facet Grid (sns.FacetGrid)


`FacetGrid` is a powerful class in Seaborn that allows you to create a grid of subplots based on the values of one or more categorical variables. It is particularly useful for visualizing relationships between multiple variables in a structured way.

### Here's a brief overview of what `FacetGrid` does:

- It creates a grid of subplots from a DataFrame.
- It allows you to map different variables to the rows, columns, and hue of the grid.
- It enables you to apply a plotting function to each subset of the data.

### Key Features:

- **Faceting:** You can create subplots based on the values of one or more categorical variables using the `row`, `col`, and `hue` parameters.
- **Customization:** You can customize the appearance and behavior of the grid using various parameters such as `margin_titles`, `despine`, and `aspect`.
- **Plot Mapping:** You can map a plotting function to the grid using the `map` or `map_dataframe` methods, allowing for complex vode examples below for more information.

```python
df = pd.read_csv("university_enrollment_(Mini-project-01).csv")
g = sns.FacetGrid(df, col = "course_type", hue = "pre_requirement", )
g.map_dataframe(sns.violinplot, x = "enrollment_count")
plt.legend()
```
