# Artificial Neural Network in TensorFlow
An Artificial Neural Network (ANN) in TensorFlow is a model based on how information processing takes place in biological neural networks residing in the human brain. In simple terms, an ANN is an interconnected set of nodes, or neurons, that operate by learning from patterns in data and making predictions or decisions.

## Key Concepts:
1. **Layers:**
   - **Input Layer:** The first layer that receives the input data.
   - **Hidden Layers:** Intermediate layers where the actual computation occurs. These layers apply transformations to the data through a series of weighted connections and activation functions.
   - **Output Layer:** The final layer that produces the output, which could be a prediction or classification result.
2. **Neurons:**
   - Each neuron receives input, processes it through a weighted sum, applies an activation function, and passes the result to the next layer. The weights are adjusted during training to minimize the error in predictions.
3. **Activation Functions:**
   - These are mathematical functions applied to the output of each neuron to introduce non-linearity into the model, enabling it to learn complex patterns. Common activation functions include ReLU (Rectified Linear Unit), Sigmoid, and Tanh.
3. **Training:**
   - Training an ANN involves feeding it data, comparing the predictions with the actual results, and adjusting the weights through a process called backpropagation. The goal is to minimize the difference between the predicted and actual outcomes, usually quantified by a loss function.
4. **TensorFlow:**
   - TensorFlow is a popular open-source framework developed by Google that provides tools for building and training ANNs. It allows you to define the structure of the network, specify how data flows through it, and perform the necessary calculations to train the model. TensorFlow handles the complexities of matrix operations, optimizations, and parallel computing, making it easier to implement and scale neural networks.

# Activation
Activation functions play a crucial role in neural networks by introducing non-linearity into the model, allowing it to learn complex patterns. Here's a brief overview of some commonly used activation functions:
1. **Sigmoid:**
    - **Formula:** $$
\sigma(x) = \frac{1}{1 + e^{-x}}
$$
    - Output Range: 0 to 1
    - Description: The sigmoid function maps any real-valued number into a value between 0 and 1, making it useful for binary classification problems. The output can be interpreted as a probability. However, it suffers from the vanishing gradient problem, where gradients become very small, slowing down the learning process for deep networks.
2. ReLU (Rectified Linear Unit):
    - **Formula:** $$
\text{ReLU}(x) = \max(0, x)
$$
    - Output Range: 0 to infinity
    - Description: ReLU is one of the most popular activation functions in deep learning. It outputs the input directly if it is positive; otherwise, it returns zero. ReLU helps to alleviate the vanishing gradient problem and is computationally efficient. However, it can suffer from the "dying ReLU" problem, where neurons can sometimes become inactive and only output zero.
4. Softmax:
    - **Formula:** $$
\text{Softmax}(x_i) = \frac{e^{x_i}}{\sum_{j} e^{x_j}}
$$
    - Output Range: 0 to 1 (sum of outputs equals 1)
    - Description: Softmax is typically used in the output layer of a neural network for multi-class classification problems. It converts the raw output scores (logits) of a neural network into probabilities that sum to 1. This allows for clear probabilistic interpretation of the network's predictions.
6. Linear:
    - **Formula:** $$
\text{Linear}(x) = x
$$
    - Output Range: Negative to positive values
    - Description: The linear activation function is essentially the identity function, meaning the output is equal to the input. It is used in cases where a linear relationship is sufficient, such as in regression tasks. However, without non-linearity, a network with linear activation functions in all layers behaves like a single-layer linear model, limiting its ability to learn complex patterns.

# Loss Functions

Loss functions measure how well a neural network is performing by quantifying the difference between the predicted output and the actual output. Here's a brief overview of some commonly used loss functions:

1. **Mean Squared Error (MSE):**
    - **Formula:** 
    $$
    \text{MSE} = \frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2
    $$
    - **Output Range:** 0 to infinity
    - **Description:** MSE calculates the average squared difference between the predicted and actual values. It's commonly used for regression tasks. The goal is to minimize the MSE, meaning the predictions should be as close as possible to the actual values.

2. **Binary Cross-Entropy (Log Loss):**
    - **Formula:** 
    $$
    \text{Binary Cross-Entropy} = -\frac{1}{n} \sum_{i=1}^{n} [y_i \log(\hat{y}_i) + (1 - y_i) \log(1 - \hat{y}_i)]
    $$
    - **Output Range:** 0 to infinity
    - **Description:** Binary cross-entropy is used for binary classification tasks. It measures the dissimilarity between the predicted probability and the actual binary label (0 or 1). A lower value indicates a better model.

3. **Categorical Cross-Entropy:**
    - **Formula:** 
    $$
    \text{Categorical Cross-Entropy} = -\sum_{i=1}^{n} y_i \log(\hat{y}_i)
    $$
    - **Output Range:** 0 to infinity
    - **Description:** Categorical cross-entropy is used for multi-class classification problems. It calculates the loss between the true label and the predicted probabilities for each class. The goal is to minimize the cross-entropy, improving the model’s predictions.

4. **Hinge Loss:**
    - **Formula:** 
    $$
    \text{Hinge Loss} = \frac{1}{n} \sum_{i=1}^{n} \max(0, 1 - y_i \cdot \hat{y}_i)
    $$
    - **Output Range:** 0 to infinity
    - **Description:** Hinge loss is used primarily for "maximum-margin" classification, most notably for support vector machines. It ensures that predictions not only get classified correctly but also with a margin of safety.

5. **Huber Loss:**
    - **Formula:** 
    $$
    \text{Huber Loss} = 
    \begin{cases} 
    \frac{1}{2}(y_i - \hat{y}_i)^2 & \text{for } |y_i - \hat{y}_i| \leq \delta, \\
    \delta \cdot (|y_i - \hat{y}_i| - \frac{1}{2}\delta) & \text{otherwise}
    \end{cases}
    $$
    - **Output Range:** 0 to infinity
    - **Description:** Huber loss is a combination of MSE and MAE (Mean Absolute Error) and is less sensitive to outliers. It behaves like MSE for small errors and like MAE for large errors, making it robust for regression tasks.

6. **Kullback-Leibler Divergence (KL Divergence):**
    - **Formula:** 
    $$
    D_{\text{KL}}(P || Q) = \sum_{i} P(i) \log \frac{P(i)}{Q(i)}
    $$
    - **Output Range:** 0 to infinity
    - **Description:** KL Divergence is used to measure how one probability distribution diverges from a second, expected probability distribution. It’s often used in vari- ational autoencoders and other probabilistic models.

# Importing Dependencies

```python
import pandas as pd
import numpy as np
from tensorflow.keras.layers import Dense, Dropout, Input
from tensorflow.keras.models import Sequential
from tensorflow.keras.optimizers import SGD, RMSprop, Adam
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from tensorflow.keras.utils import plot_model
```

```python
x = df.drop('Gender', axis = 1)
y = df['Gender']
```

- Here i defined the function to do all in just one step, in the parameters of the function you have to inset teh optimizer, learning rate, epoches, batch size

```python
def dnn_model(x, y, opti, lr, epo, bat):
    x_train, x_temp, y_train, y_temp = train_test_split(x, y, test_size=0.2, random_state=42, shuffle=True)
    x_test, x_val, y_test, y_val = train_test_split(x_temp, y_temp, test_size=0.2, random_state=42, shuffle=True)

    model = Sequential([
    Input(shape=(x_train.shape[1],)),
    Dense(128, activation = 'relu'),
    Dense(64, activation = 'relu'),
    Dense(y_train.nunique(), activation = 'sigmoid')
    ])

    model.summary()

    model.compile(optimizer = opti(learning_rate = lr), loss = 'sparse_categorical_crossentropy', metrics = ['accuracy'])

    # Train the model and capture the history
    modelFit = model.fit(x_train, y_train, epochs=epo, batch_size=bat, validation_data=(x_val, y_val))

    # Evaluate the model on the test set
    loss, accuracy = model.evaluate(x_test, y_test)
    
    print(f'Test Loss: {loss}')
    print(f'Test Accuracy: {accuracy}')

    # Plot the training history (accuracy and loss)
    plt.figure(figsize=(14, 5))
    
    # Plot accuracy
    plt.subplot(1, 2, 1)
    plt.plot(modelFit.history['accuracy'], label='Training Accuracy')
    plt.plot(modelFit.history['val_accuracy'], label='Validation Accuracy')
    plt.title('Model Accuracy')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend()
    
    # Plot loss
    plt.subplot(1, 2, 2)
    plt.plot(modelFit.history['loss'], label='Training Loss')
    plt.plot(modelFit.history['val_loss'], label='Validation Loss')
    plt.title('Model Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()
    
    plt.show()

dnn_model(x, y, Adam, 0.01, 500, 32)
```
